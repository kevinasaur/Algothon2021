```python
import pandas as pd
import sklearn
import numpy as np
from sklearn import preprocessing

data = np.loadtxt('prices250.txt')

prices = data[:,:]
```

```python
data_normaliser = preprocessing.MinMaxScaler()
prices_normalised = data_normaliser.fit_transform(prices)
```

```python
prices_normalised
```

```python
average_mean = prices.mean(axis=1)
average_mean_normalised = prices_normalised.mean(axis=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(average_mean)
plt.show()
plt.plot(average_mean_normalised)
plt.show()
```
